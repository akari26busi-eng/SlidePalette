```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from time import sleep
import csv
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# csvファイルに追記する関数
def appendcsv(filename, line):
    with open(filename, 'a', newline='\n', encoding='utf_8') as f:
        writer = csv.writer(f)
        writer.writerow(line)

# Chromeドライバの設定
chrome_options = Options()
chrome_options.add_argument('--headless')  # ヘッドレスモード
chrome_options.add_argument('--incognito') # シークレットモードで起動

driverpath = './chromedriver-win64/chromedriver.exe'

driver = webdriver.Chrome(driverpath)
```

```python
# csvファイルに追記する関数
def appendcsv(filename, line):
    with open(filename, 'a', newline='\n', encoding='utf_8') as f:
        writer = csv.writer(f)
        writer.writerow(line)
```

```python
base_url='https://www.ted-ja.com/2019/'
driver.get(base_url)
sleep(5)


# 保存するファイル
filename = 'ted.csv'

# アコーディオン展開用のボタンをクリック
accordion_buttons = driver.find_elements(By.CSS_SELECTOR, "span.zippy")
for button in accordion_buttons[2:]:
    try:
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable(button))
        button.click()  # アコーディオンボタンをクリックして展開
        sleep(2)
    except Exception as e:
        print(f"ボタンをクリックできませんでした: {e}")

# 展開後、各投稿を取得
posts = driver.find_elements(By.CSS_SELECTOR, "ul.posts li a")
for post in posts:
    try:
        title = post.text  # 投稿のタイトルを取得
        if '｜' in title:
            title = title.split('｜')[0]
        if '|' in title:
            title = title.split('|')[0]
        if '—' in title:
            title = title.split('—')[0]
        elif '―' in title:
            title = title.split('―')[0]
        elif '-' in title:
            title = title.split('-')[0]
        if ':' in title:
            title = title.split(':')[1]
        elif '：' in title:
            title = title.split('：')[1]
        if '／' in title:
            title = title.split('／')[1]
        if 'タイトルなし' in title:
            continue
        if '(TEDx' in title:
            title = title.split('(TEDx')[0]
        if '「' in title:
            title = title.split('「')[1]
        if '『' in title:
            title = title.split('『')[1]
            
            
        title = title.replace(' ', '')
        
        appendcsv(filename, [title])  # CSVに保存

    except Exception as e:
        print(f"投稿の取得に失敗しました: {e}")



# WebDriverを終了
driver.quit()
```
