```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from time import sleep
```

```python
import csv
import pandas as pd
import re
```

```python
# Chromeドライバの設定
chrome_options = Options()
chrome_options.add_argument('--headless')  # ヘッドレスモード
chrome_options.add_argument('--incognito') # シークレットモードで起動

driverpath = './chromedriver-win64/chromedriver.exe'

driver = webdriver.Chrome(driverpath)
```

```python
# 無彩色を判定する関数
def is_achromatic(rgb):
    r, g, b = map(int, rgb)
    return r == g == b or (abs(r - g) < 15 and abs(g - b) < 15 and abs(r - b) < 15)
```

```python
# csvファイルに追記する関数
def appendcsv(filename, line):
    with open(filename, 'a', newline='\n', encoding='shift_jis') as f:
        writer = csv.writer(f)
        writer.writerow(line)
```

```python
base_url = "https://iro-color.com/colorchart/tint/seasonword-autumn.html"
```

```python
driver.get(base_url)
sleep(5)

# 配色キーワードの辞書
keywords_url = {}

# 保存するファイル
filename = 'keywords_url.csv'

# 配色キーワードのリストを取得
list_keywords_url = driver.find_elements(By.CSS_SELECTOR, "ul.article_list li a")


# 各リンクを辞書に格納
for item in list_keywords_url:
    text = item.text
    url = item.get_attribute("href")

    if "ホームページの配色を提案します" == text:
        break
    elif ("五十音のイメージ配色" == text) or ("月" in text) or ("時間" in text) or ("誕生石" in text) or ("季語" in text) or ("パワーストーン" in text) or ("タロット" in text) or ("曜日" in text):
        print(text + 'は飛ばす\n')
        continue


    if "カラーチャート" in text:
        # 「の」より前をスライス
        key = text.split("の")[0]
    elif "を表現" in text:
        # 「を」より前をスライス
        key = text.split("を")[0]    
    elif "に関する" in text:
        # 「に」より前をスライス
        key = text.split("に")[0]  
    elif "鳥" in text:
        key = '鳥'
    elif "イメージ配色" in text:
        # 「の」より前をスライス
        key = text.split("の")[0]   

    keywords_url[key] = url
    appendcsv(filename, [key, url])
```

```python
# 単語とURLの辞書
word_url = {}

# 保存するファイル
filename = 'word_url.csv'

df = pd.read_csv("keywords_url.csv", encoding='shift_jis')

for url in df['url']:
    driver.get(url)  # URLにアクセス
    sleep(10)  # 必要に応じてページの読み込みを待つ
    # ページ内のターゲット要素を取得
    list_word_url = driver.find_elements(By.CSS_SELECTOR, "ul.contents_tag li a")

    for item in list_word_url:
        try:
            word = item.text
            url = item.get_attribute("href")
            
            if word not in word_url:
                # 辞書に同じ単語がないとき、単語とURLを追加
                word_url[word] = url
                appendcsv(filename, [word, url])

        except Exception as e:
            print(f"エラー: {url} で単語を取得できませんでした: {e}")

print(word_url)
```

```python
# ファイル名
filename = "color_data.csv"

df = pd.read_csv("word_url.csv", encoding='shift_jis')


for word, url in zip(df['word'][631:], df['url'][631:]):
    driver.get(url)  # URLにアクセス
    sleep(5)

    # HTMLを取得
    html = driver.page_source

    # HEXとRGBを正規表現で取得
    # パターンマッチするものをリストで取得
    hex_pattern = r"HEX：(#\w{6})"
    rgb_pattern = r"RGB：R(\d+)／G(\d+)／B(\d+)"
    
    hex_matches = re.findall(hex_pattern, html)
    rgb_matches = re.findall(rgb_pattern, html)

    print(hex_matches)
    print(rgb_matches)

    colorcode_re = [(hex, rgb) for hex, rgb in zip(hex_matches, rgb_matches) if is_achromatic(rgb) == False]
    print(colorcode_re)
    for hex, rgb in colorcode_re:
        r, g, b = map(int, rgb)

        # 同じ色を登録していないか確認
        # 同じ色しかない場合は一番後ろの有彩色を登録
        if colorcode_re[len(colorcode_re)-1][0] == hex:
            print([word, hex, r, g, b])  
            appendcsv(filename, [word, hex, r, g, b])
            colorcode.append(hex)
            break
        elif hex not in colorcode:
            print([word, hex, r, g, b])  
            appendcsv(filename, [word, hex, r, g, b])
            colorcode.append(hex)
            break

# ドライバーを終了
driver.quit()
```
