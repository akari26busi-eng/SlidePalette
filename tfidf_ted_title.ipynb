```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(analyzer=lambda x: x)
X_vec = vectorizer.fit_transform(X)

# print(X)

# 各文の重要語を抽出
feature_names = vectorizer.get_feature_names_out()
for doc_index, row in enumerate(X_vec):
    sorted_indices = row.toarray().argsort()[0, -2:][::-1]  # スコア上位2語を抽出
    predicted_words = [feature_names[idx] for idx in sorted_indices]
    print(f"文 {doc_index + 1} の重要語: {predicted_words}")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# 評価用
precision_list = []
recall_list = []
f1_list = []

for doc_index, row in enumerate(X_vec):
    # 予測重要語: X の中から上位 2 単語（ただしインデックスで指定）
    sorted_indices = row.toarray().argsort()[0, -2:][::-1]  # スコア上位2語を抽出
    predicted_words = [feature_names[idx] for idx in sorted_indices]

    # 正解ラベル（重要語のインデックス）
    true_indices = y[doc_index]
    # print(true_indices)
    # print(X[doc_index])
    # doc_length = len(X[doc_index])
    # for i in true_indices:
    #     if i >= doc_length:
    #         print(f"警告{doc_index}: インデックス {i} は、長さ {doc_length} のドキュメントの範囲外です")
    
    # 正解ラベル（重要語）
    true_words = [X[doc_index][i] for i in true_indices]

    # Precision, Recall, F1の計算
    tp = len(set(predicted_words) & set(true_words))
    fp = len(set(predicted_words) - set(true_words))
    fn = len(set(true_words) - set(predicted_words))

    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1 = (2 * tp) / (2 * tp + fp + fn) if (2 * tp + fp + fn) > 0 else 0

    precision_list.append(precision)
    recall_list.append(recall)
    f1_list.append(f1)

    print(f"文 {doc_index + 1} の予測重要語: {predicted_words}")
    print(f"文 {doc_index + 1} の正解重要語: {true_words}")
    print(f"Precision: {precision}  Recall: {recall}  F1: {f1}")

# スコアの平均値を出力
print(f"平均 Precision: {sum(precision_list) / len(precision_list):.4f}")
print(f"平均 Recall: {sum(recall_list) / len(recall_list):.4f}")
print(f"平均 F1スコア: {sum(f1_list) / len(f1_list):.4f}")
```
